```python
df.drop(columns=['PostedDate'],inplace=True)
```

```python
df['kmDriven']=df['kmDriven'].apply(lambda x:str(x).split(" ")[0])
df['kmDriven']=df['kmDriven'].str.replace(",","")
```

```python
df['kmDriven']=df['kmDriven'].astype("float64")
```

```python
df.rename(columns={"AskPrice":"Price"},inplace=True)
```

```python
df['Price']=df['Price'].apply(lambda x:str(x).split(" ")[1])
df['Price']=df['Price'].str.replace(",","")
```

```python
df['Price']=df['Price'].astype("float64")
```

```python
def remove_outlier(df,col):
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    iqr=Q3-Q1
    df.iloc[(df[col]<(Q1-1.5*iqr))|(df[col]>(Q3+1.5*iqr))]=np.nan
    df.dropna(inplace=True)
    return df
```

```python
df['Price'].count()
```

```python
df=remove_outlier(df,'Price')
```
